## Quantos foram por cada ano ?

```python
emprestimos_data = pd.DataFrame(emprestimos_completo['data_emprestimo'].value_counts()).reset_index()
emprestimos_data.columns = ['data','quantidade']
emprestimos_data['data'] = pd.to_datetime(emprestimos_data['data'])
emprestimos_data
```

```python
emprestimos_por_ano = emprestimos_data.groupby(by=emprestimos_data.data.dt.year).sum()
emprestimos_por_ano
```

```python
emprestimos_por_ano['percentual'] = round((emprestimos_por_ano.quantidade / emprestimos_por_ano.quantidade.sum()) * 100, 2)
emprestimos_por_ano
```

## Quantos foram por cada mês ?

```python
emprestimos_por_mes = emprestimos_data.groupby(by=emprestimos_data.data.dt.month).sum()
emprestimos_por_mes
```

```python
emprestimos_por_mes['percentual'] = round((emprestimos_por_mes.quantidade / emprestimos_por_mes.quantidade.sum()) * 100, 2)
emprestimos_por_mes
```

## Quantos foram por cada ano e mês ?


```python
emprestimos_por_ano_e_mes = emprestimos_data.groupby(by=[emprestimos_data.data.dt.year,emprestimos_data.data.dt.month]).sum()
emprestimos_por_ano_e_mes.index.names = ['ano','mes']
emprestimos_por_ano_e_mes
```
